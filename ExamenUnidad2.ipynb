```python
import pandas as pd
import io

# Load the uploaded file into a DataFrame
df = pd.read_csv(io.BytesIO(uploaded['items.csv']))
print("Original DataFrame (df):")
display(df)

# Create the 'importe' column using eval()
df.eval('importe = precio * cantidad', inplace=True)
print("DataFrame after adding 'importe' column:")
display(df)

# Filter for categoria == 'A' using query() and store in dfa
dfa = df.query("categoria == 'A'")
print("Filtered DataFrame (dfa) for categoria == 'A':")
display(dfa)

# Calculate the average of 'importe' per category in the original DataFrame df
average_importe_by_category = df.groupby('categoria')['importe'].mean()
print("Average 'importe' by category (from original DataFrame):")
display(average_importe_by_category)

# Get the average importe for category 'A' and round to 3 decimal places
average_importe_A = round(average_importe_by_category['A'], 3)

print(f"\nEl promedio de importe de la categoría A (en el DataFrame original) es: {average_importe_A}")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

# Load ordenes.csv into dfo
dfo = pd.read_csv(io.BytesIO(uploaded['ordenes.csv']))
print("DataFrame de Órdenes (dfo):")
display(dfo)

# Load clientes.csv into dfc
dfc = pd.read_csv(io.BytesIO(uploaded['clientes.csv']))
print("DataFrame de Clientes (dfc):")
display(dfc)

# Perform a left merge of dfc with dfo by cliente_id, keeping all rows of dfc
merged_df = pd.merge(dfc, dfo, on='cliente_id', how='left')
print("DataFrame después del left merge (merged_df):")
display(merged_df)

# (a) Calculate total number of rows of the merged_df
total_rows_merged = len(merged_df)
print(f"\n(a) Número total de filas del resultado del left merge: {total_rows_merged}")

# (b) Calculate total amount per client and replace NaN with 0
monto_total_por_cliente = merged_df.groupby('cliente_id')['monto'].sum().fillna(0)
print("\n(b) Monto total por cliente (NaNs reemplazados por 0):")
display(monto_total_por_cliente)

# Get the total amount for cliente_id=1
monto_cliente_1 = monto_total_por_cliente.loc[1]
print(f"\nEl monto total por cliente con cliente_id=1 es: {monto_cliente_1}")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

# Load the uploaded file into a DataFrame
df_sensores = pd.read_csv(io.BytesIO(uploaded['sensores.csv']))
print("Original DataFrame (df_sensores):")
display(df_sensores)

# Convert 'ts' to datetime, handling mixed formats and coercing errors
df_sensores['ts'] = pd.to_datetime(df_sensores['ts'], format='mixed', errors='coerce')
print("DataFrame after converting 'ts' to datetime:")
display(df_sensores)

# Calculate median for 'temperatura' and mean for 'humedad'
median_temperatura = df_sensores['temperatura'].median()
mean_humedad = df_sensores['humedad'].mean()

# Replace missing values
df_sensores['temperatura'].fillna(median_temperatura, inplace=True)
df_sensores['humedad'].fillna(mean_humedad, inplace=True)
print("DataFrame after imputing missing values:")
display(df_sensores)

# (a) Count NaT values remaining in 'ts'
nat_ts_count = df_sensores['ts'].isna().sum()
print(f"\n(a) Número de NaT restantes en 'ts': {nat_ts_count}")

# (b) Final median of 'temperatura' after imputation
final_median_temperatura = round(df_sensores['temperatura'].median(), 1)
print(f"(b) Mediana final de 'temperatura' : {final_median_temperatura}")

# (c) Final mean of 'humedad' after imputation
final_mean_humedad = round(df_sensores['humedad'].mean(), 3)
print(f"(c) Media final de 'humedad' : {final_mean_humedad}")
```
